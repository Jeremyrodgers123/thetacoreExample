```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
```

```python
credit = pd.read_csv('Credit.csv')
```

```python
credit = credit[['Income', 'Limit', 'Rating', 'Cards', 'Age', 'Education', 'Gender', 'Student', 'Married','Ethnicity', 'Balance']]
```

```python
credit
```

```python
print(credit.describe())
```

```python
credit.corr()
```

```python
pd.plotting.scatter_matrix(credit)
```

```python
limitModel = sm.ols(data=credit, formula="Limit ~ Rating").fit()
display(limitModel.summary())
```

```python
credit["Gender_num"] = credit['Gender'].map({"Female" : 1, "Male" : 0})
credit["Married_num"] = credit['Married'].map({"Yes" : 1, "No" : 0})
credit["Student_num"] = credit['Student'].map({"Yes" : 1, "No" : 0})
```

```python
dummies = pd.get_dummies(credit["Ethnicity"])
frames = [credit, dummies]
credit = pd.concat(frames, axis = 1)
credit
```

```python
credit
```

```python
formula = "Limit ~ Income + Cards + Age + Education + Gender_num + Student_num + Married_num + Caucasian"
```

```python
limitModel = sm.ols(data=credit, formula= formula).fit()
```

```python
display(limitModel.summary())
```
